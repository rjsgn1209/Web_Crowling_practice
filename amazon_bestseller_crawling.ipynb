```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome()
driver.get("https://www.amazon.com/bestsellers")
```

```python
elem = driver.find_element_by_link_text("Appliances")
elem.click()
```

```python
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
items = soup.select("div.a-section")
print(items)
```

```python
elem2 = driver.find_element_by_link_text("2")
                                      
elem2.click()
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
items += soup.select("div.a-section")
```

```python
for item in items:
    rank = item.select("span.zg-badge-text")
    if rank:
        print(rank[0].text, end=" ")
    title = item.select("div.p13n-sc-truncated")
    if title:
        print(title[0].text)
    score = item.select("a.a-size-small")
    if score:
        print(score[0].text, end=" ")
    price = item.select("span.p13n-sc-price")
    if price:
        print(price[0].text)
        print("=" * 50)
```
